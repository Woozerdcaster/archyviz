## What is uv?  
uv is Astral’s fast Python package & project manager. It can manage Python versions, create virtual environments, handle dependencies via `pyproject.toml`, and produce reproducible lockfiles.  
**References:** [Official installation guide](https://docs.astral.sh/uv/getting-started/installation/), [Using environments](https://docs.astral.sh/uv/pip/environments/), [Working on projects](https://docs.astral.sh/uv/guides/projects/), [PyPI overview](https://pypi.org/project/uv/).

We are only going to use it in a very limited capacity, as a way to create a virtual environment and to install the necessary packages needed for our course. You are encouraged to explore its capabilities further by visiting the [uv project](https://docs.astral.sh/uv/) website.

There are potentially many different ways in which you can use `uv`, more than what we are going to cover here. 


## Download & Install uv 

We’ll download the official installer scripts and run them locally.

### macOS (Terminal: bash/zsh)
1. Open the official installation page: <https://docs.astral.sh/uv/getting-started/installation/>. 
2. Go to the *Standalone installer* section.
3. Copy the installation script.
4. Open a Terminal and paste either of the following:
    - Using `sh`:
    ```bash
    > curl -LsSf https://astral.sh/uv/install.sh | sh
    ```
    - OR using `wget`:
    ```bash
    > wget -qO- https://astral.sh/uv/install.sh | sh
    ```


### Windows (PowerShell or Command Prompt)
1. Open the same installation page.
2. Go to the *Standalone installer* section.
3. Copy the installation script.
4. Open a PowerShell or Command Prompt and paste:

```powershell
> powershell -ExecutionPolicy ByPass -c "irm https://astral.sh/uv/install.ps1 | iex"
```

### Upgrade UV

In the same terminal or command prompt type
```bash
uv self update
```

## Creating an environment

When developing a project, you will want to generate a virtual environment that stores and/or makes reference to all that you need to develop the project. Within this virtual environment, you may want to make reference to a specific version of Python and a list of packages. These packages have to be compatible with whatever version of Python you are planning to use. `uv` helps you to build such an environment and to populate it with the right packages so that they are all compatible. 

```{caution}
Keep in mind that right now, we are **NOT** developing a **project**. We are going to use `uv` to create a virtual environment with a specific version of Python. We will then use this virtual environment to run notebooks.
```

### *Preliminary check*   

`uv` has been designed to use Python versions that are on your machine. If you want to develop some project with a version of Python that is not on your machine, `uv` can help you install it.

Installing any version of Python on a machine can be done in different ways but here we will just stick to one. Before we start, let us see what Python installations are already available.  Type,

```bash
uv python list
```

`uv` can detect all installations, even those that it does not manage. Eventually, you may want to spend some time learning about the different options that are available with [`uv python list`](https://docs.astral.sh/uv/reference/cli/#uv-python-list).

**Reference:** [Using environments → Activation & deactivation](https://docs.astral.sh/uv/pip/environments/)

Now, let us imagine that for our future virtual environment we want to use the Python version 3.14. **If this version is not installed on your machine** (or not managed by `uv`), we will type,
```bash
uv python install 3.14
```
to install it.

### Creating **ONLY** a virtual environment    

Create a folder where you are going to put all of your Python code (mine is `C:\Python`). Within this folder, you are going to want to create a folder for each project you are going to work on. This folder will contain your virtual environment. You can create these folders yourself or you can use `uv` to generate it with the following steps. Type,   

```bash
uv venv archyviz --python 3.14
```
This will generate a new folder (`archyviz`) that will contain several files amongst them a `.venv` directory. Check that `uv` created such a folder by typing on your computer prompt

```powershell
> dir
```
in Windows OR
```bash
> ls
```
in mac or linux.

### Activate the environment

If you want to activate your environment you will need to type,

```bash
# macOS/Linux
source <name of environment>/bin/activate

# Windows (PowerShell)
<name of environment>\Scripts\activate
```
To deactivate,
```bash
# exit (both)
deactivate
```
**Reference:** [Using environments → Activation & deactivation](https://docs.astral.sh/uv/pip/environments/)

Now your enironment current does not contain anything.

### Installing packages    

There are many ways to install packages. You can use `uv`to install them one at a time or if you had already a `uv`project you could use the `.toml` associated with it (more later). Right now we are going to use a simple text file, named `requirements.txt`that contains all of the packages we want to use at the moment.

 #### Installing using `requirements.txt`

Download the project's `requirements.txt` file into your machine. If you want to take a look at what is inside, open it with a simple text editor. Copy this file into the `/archyviz` folder you created and type the following,
 
```bash
uv pip install -r requirements.txt
```    

This will download and install all of the packages that we need. If you want to make sure that only those files are installed an no extras, type 

```bash
uv pip sync -r requirements.txt   # exact match (removes extras)
```


### Installing using `pyproject.toml`. `uv.lock` and `python.version`

You may have already a project with its environment that was created using `uv`. If you want to reproduce exactly that environment.

- Create a new folder.
- Copy all of the files `pyproject.toml`. `uv.lock` and `python.version` into this folder.
- Next type,
```bash
uv venv <name of environment>
uv sync --frozen
```

This will install everything exactly as you had it in your projects.

Once you have created the virtual environment and installed the packages, you can run the notebooks by starting Jupyter (e.g., `uv run jupyter lab`) or by using `uv run` to execute commands in the project environment.


## Create a project **from scratch** 

1) Open a terminal or command prompt and initialize the project (**set the verion of Python you want to use**):

```bash
uv init example_project --python 3.12 # setting project to python 3.12
```
Example layout `example_project/`:

```
example_project/
├─ .gitignore
├─ python-version
├─ main.py
├─ pyproject.toml
├─ README.md
└─ .venv/         (created by `uv venv`, runnin `main.py`or installing some package)
```
This creates `pyproject.toml`, `.python-version`, `README.md`, `.gitignore`, and a starter script (e.g., `main.py`).  
**Reference:** [Working on projects → `uv init` & project files](https://docs.astral.sh/uv/guides/projects/)

2) Add packages. You can add packages using `uv add`
```bash
uv add pandas jupyterlab 
```
The moment you add any package, `uv` will create a virtual environement with the same name as the project

This updates `pyproject.toml` automatically.  
**Reference:** [Working on projects → Managing dependencies](https://docs.astral.sh/uv/guides/projects/)

3) **Best way to set up the environment using `pyproject.toml` (recommended):**

- Treat `pyproject.toml` as the **single source of truth**. Use `uv add`/`uv remove` to manage dependencies rather than editing the file manually.
- **Lock** dependencies and **sync** the environment:

```bash
uv lock
uv sync
```

This produces `uv.lock` for reproducible installs across machines. Commit `pyproject.toml`, `.python-version`, and `uv.lock` to version control.  
**Reference:** [Working on projects → Lockfile & sync](https://docs.astral.sh/uv/guides/projects/)

4) Run inside the managed environment (no manual activation needed for most tasks):

```bash
uv run python main.py
```

Manual activation is still available:

```bash
# macOS/Linux
source .venv/bin/activate

# Windows (PowerShell)
.\.venv\Scripts\activate
```

5) If you later need a different Python for the project:

```bash
uv python install 3.11
uv python pin 3.11
uv sync
```

**Reference:** [Cheat sheet → python install/pin](https://ricketyrick.github.io/uv_cheat_sheet/uv-cheat-sheet.html)


## Most‑Used `uv` Commands — Quick Reference

Below are common commands plus a brief description of what each does.

```bash
# Inspect installed packages (optional)
uv pip list
```
> Lists packages in current environment.
<br/>

<br/>

```bash
# Check uv version
uv --version
```
> Prints uv’s version to verify installation.  
> **Ref:** [PyPI](https://pypi.org/project/uv/)   
<br/>

<br/>

```bash
# Create a virtual environment (default .venv), specifying Python version
uv venv --python 3.12
```
> Creates `.venv/` using Python 3.12 (downloads the interpreter if missing).  
> **Ref:** [Using environments](https://docs.astral.sh/uv/pip/environments/)
<br/>

<br/>

```bash
# Activate/deactivate the environment
source .venv/bin/activate         # macOS/Linux
.\.venv\Scripts\activate    # Windows (PowerShell)
deactivate
```
> Activates/deactivates the venv for ad‑hoc Python usage.  
> **Ref:** [Using environments](https://docs.astral.sh/uv/pip/environments/)
<br/>

<br/>

```bash
# Initialize a new project (scaffold) and pin Python
uv init --python 3.12
```
> Creates `pyproject.toml`, `.python-version`, starter files; pins Python for the project.  
> **Ref:** [Working on projects](https://docs.astral.sh/uv/guides/projects/)
<br/>

<br/>

```bash
# Add pacakges to pyproject.toml
uv add pandas
uv add matplotlib
```
> Updates `pyproject.toml` and installs dependencies into the project environment.  
> **Ref:** [Working on projects → Managing dependencies](https://docs.astral.sh/uv/guides/projects/)
<br/>

<br/>

```bash
# Create/refresh lockfile and install exactly what's locked
uv lock
uv sync
```
> Generates `uv.lock` and syncs the environment for reproducible installs.  
> **Ref:** [Working on projects → Lockfile & sync](https://docs.astral.sh/uv/guides/projects/)
<br/>

<br/>

```bash
# Install packages from requirements.txt (pip-compatible interface)
uv pip install -r requirements.txt
uv pip sync -r requirements.txt
```
> Fast install and exact sync of environment to match `requirements.txt`.  
> **Refs:** [Stack Overflow example](https://stackoverflow.com/questions/79344035/how-to-add-requirements-txt-to-uv-environment), [Cheat sheet](https://ricketyrick.github.io/uv_cheat_sheet/uv-cheat-sheet.html)
<br/>

<br/>

```bash
# Run commands inside the project environment without manual activation
uv run python main.py
```
> Executes `python main.py` within the managed environment. Works for tools, tests, etc.  
> **Ref:** [Working on projects → `uv run`](https://docs.astral.sh/uv/guides/projects/)
<br/>

<br/>

```bash
# Ensure a Python version exists and pin it for this project
uv python install 3.11
uv python pin 3.11
```
> Downloads the requested interpreter if needed, and writes `.python-version` for the project.  
> **Ref:** [Cheat sheet → python install/pin](https://ricketyrick.github.io/uv_cheat_sheet/uv-cheat-sheet.html)
```
