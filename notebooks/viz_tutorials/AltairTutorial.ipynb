## 1. Introduction <br>
**Altair** is a declarative statistical visualization library for Python built on the grammar of the Vega ecosystem. Altair is designed for information and statistical visualization, interactive exploratory analysis, and lightweight dashboarding. It allows users to create clear, expressive, and interactive visualizations using a readable syntax.
<br>
<br>
Altair follows a *grammar of graphics* approach. Rather than specifying how each axis or element is drawn, you describe what you want to render: <br>
- The data itself <br>
- The variables to visualize <br>
- How those variables are encoded <br>
- Transformations like aggregation, filtering, or grouping<br>

Altair handles the rest!<br>

### Altair is built around the Pandas dataframe
It uses a declarative syntax and assumes your data is structured in a tidy tabular format, where: <br>
- Each row corresponds to one observation <br>
- Each column corresponds to one variable <br>
- Each cell = one value 

**Altair does not depend on Matplotlib**<br>
Unlike Matplolib-based tools that require multiple formatting commands, Altair uses three main components that encompass many visualization properties:
- `alt.Chart()`: the chart is the fundamental object to Altair. It takes a dataframe as a single argument.
- `.mark_*()`: the mark attribute of the chart object, used to define the type of visual representation for your data and specify how it should be visually encoded, e.g.: <br>
&nbsp;&nbsp; * `.mark_point()`: <br>
&nbsp;&nbsp; * `.mark_bar()`:
- `.encode()`: is used to map data properties to visual properties, such as the axes, color, or shape.

## 3. Quick Start

First, we'll make a simple bar chart showing ethnobotanical material by Family. <br>
To make a simple bar chart, you'll specify the Chart object, the desired mark attribute (`mark_bar()`), and then encode the desired properties.

```python
# alt.Chart(dataframe).mark().encode(properties)
alt.Chart(data).mark_bar().encode(
    x = 'FS #:N',
    y = 'count(Wood ID)',
    color='Family:N',
    tooltip=[Wood ID', 'count(Wood ID']
)
```

## Encoding Types
Notice that we appended several lines above with 'N' or 'Q'. These are basic data types that are central to encoding data with Altair. When working with data from a file (like we are), it is the data type *must always* be specified explicitly in the code. <br>
|Code|Data Type|Description|
|----|---------|-----------|
|`Q`|**Quantitative**|Encoded as numerical quantity|
|`N`|**Nominal**|Encoded as name or unordered categorical data|
|`O`|**Ordinal**|Encoded as ordered categorical data|
|`T`|**Temporal**|Encoded as date/time data|



```python
# try changing the Family data type from Nominal to Ordinal to see how it changes the color scheme
alt.Chart(data).mark_bar().encode(
    x = 'FS #:N',
    y = 'count(Wood ID)',
    #color= 'Family:'
)
```

## 4. Configuration and Customization
 
Vega-Altair allows customization on three levels:
1. **Global Configuration**: Acts on the entire Chart. Global configuration can be applied through the `configure_*` command.
     -  Useful for adjusting labels, axes properties, mark properties, and setting themes. 
     -  Global configurations will affect every mark on the chart
     -  Should only be used for thematic styling - can conflict with local configuration and encoding and cause an error. 
2. **Local Configuration**: Acts on marks within the Chart. Local configuration can be applied through the `mark_*` command.
    -  Useful for adjusting the visualization of specific marks within the Chart object, such as color, size, shape, and opacity.
    -  Local configuration can be used when layering without causing an error
    -  Will override global settings
3. **Encoding Channels**: Used to set some chart properties through the `.encode()` command.
    -  Useful for mapping a property directly on a specific value or column/ row of the data frame
    -  Use is limited to a specific set of commands
    -  Will override local and global settings

We can see how these configurations can be applied and combined in the following example, which also highlights another Vega-Altair capability, quick creation of subplots and dashboards:

### Faceting Example

```python
faceted = (alt.Chart(data)
    .mark_bar() # The mark command can take different commands to change the color and style of marks: local configuration
    .encode(
        alt.X('FS #:N'),
        alt.Y('count(Wood ID):Q', title = "Count"),
       # alt.Color("Angio/gymno:N").scale(scheme = "category20b") # Using encoding to adjust color
    ).properties(
        width=300,
        height=300)
    # title = "Counts of Angiosperm and Gymnosperm in Dataset"))
          )


# Adjust labels using global configurations
faceted.configure_header(
    titleFontSize = 18,
    titleFontStyle = "heavy",
    labelFontSize = 14
)
```

## 5. Layering in Altair
Layering is a feature of this package that allows us to combine multiple charts in the same space to create structured visualizations. You can:
* Combine multiple mark types
* Overlay summary statistics on raw data and keep them visually distinct
* Display statistics alongside observations
* Combine encodings to highlight relationships <br>
<br>In Altair, layering works by defining multiple chart objects, each with its own mark type and encoding, and combining them with the layering operator `+` or `alt.layer()`

### Layering Example 
We can use layering to visualize how wood specimen weights vary across site contexts and botanical groups. Instead of overwhelming a single mark type, layering will help us combine the raw points with summary statistics and create a visual hierarchy to distinguish between individual measurements and group means.

## 6. Interactivity
Interactivity in Altair is declarative: you describe what you want to select, then how the visualization should respond. Selections can respond to different input events (like clicks and mouseovers) and once defined, they can be used to filter or modify visual properties of the chart. <br>
Selections are defined subsets of data points that the user will interact with. The three main types of selection are: <br>
* `selection_point`, which allows selection of individual points<br>
* `selection_single`, which restricts selection to one element at a time<br>
* `selection_multi`, which allows selecting multiple elements simultaneously
