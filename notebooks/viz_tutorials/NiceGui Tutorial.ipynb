# NiceGui Tutorial

# Introduction

NiceGui is an interactive visualization layout and stylizing package that allows users to create single-page applications, reactive transformations, and event systems that matplot.lib is unable to do on its own. Notably, it allows for greater user interaction with visualizations. 

NiceGui is an interactive package, the code is “auto-reloading”. Jupyter hub only allows this when running from a file. Thus, we will be using  the command `%%writefile app.py` to create a new file that we will then run separately using `!python app1.py`. However the command `!python` continuously runs, this means all code blocks underneath are unable to run. So before moving on to the next example we will need to restart the Kernel.

#### 3D Elements

```python
%%writefile app4.py

from nicegui import ui

with ui.scene().classes('w-full h-150') as scene:
    scene.axes_helper().material('red')
    scene.sphere().material('#4488ff').move(2, 2)
    scene.cylinder(1, 0.5, 2, 20).material('#ff8800', opacity=0.5).move(-2, 1)
    scene.extrusion([[0, 0], [0, 1], [1, 0.5]], 0.1).material('#ff8888').move(2, -1)

    with scene.group().move(z=2): # makes a new group and positions it at 2 on the z axis
        scene.box().move(x=2)
        scene.box().move(y=2).rotate(0.25, 0.5, 0.75)
        scene.box(wireframe=True).material('#888888').move(x=2, y=2)

    avocado = 'https://raw.githubusercontent.com/KhronosGroup/glTF-Sample-Assets/main/Models/Avocado/glTF-Binary/Avocado.glb'
    scene.gltf(avocado).scale(40).move(-2, -3, 0.5)

ui.run()
```

## Advanced Showcase/Interactive Capabilities

#### Interactive Avocado in 3D space

```python
%%writefile app5.py

from nicegui import events, ui

def handle_drag(e: events.SceneDragEventArguments):
    ui.notify(f'You dropped the box at ({e.x:.2f}, {e.y:.2f}, {e.z:.2f})') 
    # Notifies the user at which coordinates they dropped the box at

with ui.scene(width=500, height=500,
              drag_constraints='z = 1', on_drag_end=handle_drag) as scene:
    avocado = 'https://raw.githubusercontent.com/KhronosGroup/glTF-Sample-Assets/main/Models/Avocado/glTF-Binary/Avocado.glb'
    avocado = scene.gltf(avocado).scale(40).move(z=1).draggable()

ui.switch('Draggable Avocado',
          value=avocado.draggable_,
          on_change=lambda e: avocado.draggable(e.value))

ui.run()
```

#### Plotly

```python
%%writefile app6.py
from nicegui import ui

fig = {
    'data': [
        {
            'type': 'scatter',
            'name': 'Trace 1',
            'x': [1, 2, 3, 4],
            'y': [1, 2, 3, 2.5],
            'line': {'color' : 'red', 'width': 3},
        },
        {
            'type': 'scatter',
            'name': 'Trace 2',
            'x': [1, 2, 3, 4],
            'y': [1.4, 1.8, 3.8, 3.2],
            'line': {'dash' : 'dot', 'color':'blue', 'width': 3},
        },
    ],
    'layout': {
        'margin': {'l': 15, 'r': 0, 't': 0, 'b': 15},
        'plot_bgcolor': '#E5ECF6',
        'xaxis': {'gridcolor' : 'white'},
        'yaxis': {'gridcolor' : 'white'},
    }, #Layout used to specifiy axis and background visual conditions
}
ui.plotly(fig).classes('w-full h-150') # Using plotly to plot fig

ui.run()
```
