## Figure and Axes

1. (20 pts) Create a figure with two axes.
   1. The figure should have size of 4" high and 8" wide.
   2. Plot a *sine* on the left and cosine on the *right*
   3. Each figure should have a title

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))

#first sub
axes[0].plot(x, np.sin(x))
axes[0].set_title('Sin')

#second sub
axes[1].plot(x, np.cos(x))
axes[1].set_title('Cos')

#just doing a basic visual
plt.tight_layout()
plt.show()
```

3. (30 pts) Generate 200 random points and create a scatter plot where:
    - marker sizes vary randomly from 20 to 200
    - colors vary randombly and use a colormap ('inferno')
    - transparency is set constant to 0.6

```python
np.random.seed(69)

#set 'sample' size and randomize
n = 200
x = np.random.randn(n)
y = np.random.randn(n)

#randomize color and size
colors = np.random.randn(n)
sizes = np.random.uniform(20, 200, 200)
sizes = 200 * np.random.rand(n)

#establish the figure
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(x, y, s=sizes, c=colors, cmap='inferno',
                     alpha=0.6)
fig.colorbar(scatter, ax=ax, label='color value')
ax.set_title('infernal bubbles')
#do I not need a plt.show()?, why?
```

4. (30 pts) Create a scatter plot with 300 gray background points and highlight points where x > 0.5 and y > 1 in red.
Ensure the red points appear on top using zorder and have a transparency of 0.7

```python
fig, ax = plt.subplots(figsize=(7, 5))#establish figure and axes

#establish 'sample' size/randomize
x = np.random.randn(300)
y = np.random.randn(300)
#creat random pattern
ax.scatter(
    x, y,
    s=10,
    c="grey",
    alpha=0.7,
    rasterized=True,#not sure what this does (optimization??)
    zorder=1
)

#create highlight subplot
highlight = (x > 0.5) & (y > 1)
ax.scatter(
    x[highlight], y[highlight],
    s=80,
    c='crimson',
    alpha=0.9,
    zorder=3
)

plt.show
    
```
