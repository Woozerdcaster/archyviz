# Text and Annotations

Text and annotations are essential for creating informative visualizations. In matplotlib:
- **Text**: Simple text placement at specified coordinates
- **Annotations**: Text with optional arrows pointing to specific data points

````python
import matplotlib.pyplot as plt
import numpy as np

# Set style for better-looking plots
plt.style.use('default')
````

## Text Objects - `ax.text()`

### Description and Purpose

The `ax.text()` method places text at an arbitrary location on the axes. It's used for:
- Adding labels or titles at specific positions
- Annotating plots without arrows
- Creating custom text-based visualizations
- Adding notes or explanations to plots

**Main text parameters:**
- `x`: X-coordinate of the text position. 
- `y`: Y-coordinate of the text position.
- `s`: The text string to display.
- `fontsize` or `size`: Size of the font (`'xx-small'`,`'x-small'`,`'small'`,`'medium'`,`'large'`,`'x-large'`,`'xx-large'` or numeric values, eg. `12`)
- `fontweight`: Weight/boldness of the font (`'normal'`, `'bold'`,`'heavy'`,`'light'`,`'ultrabold'`, `'ultralight'` or numeric values, eg. 100
- `fontstyle`: Style of the font (`'normal'`, `'italic'`, `'oblique'`)
- `fontname`: Name of font installed in the system, eg. 'Calibri'
- `family`: Font family (`'serif'`,`'sans-serif'`, `'monospace'`, `'cursive'`, `'fantasy'`)
- `color` or `c`: Color of the text. The usual named colors (eg. `'red'`), hex colors (eg. `'#FF5733'`), RGB tuples, RGBA tuples
-`alpha`. Transparency level
- `ha` or `horizontalalignment`: Horizontal alignment (`'left'`,`'center'`, `'right'`)
- `va` or `verticalalignment`: Vertical alignment (`'top'`, `'center'`, `'bottom'`,`'baseline'`)
- `rotation`: Rotation angle in degrees of the text or `'vertical'`, `'horizontal'`
- `bbox`: Bounding box properties (dictionary)
  - Options (dictionary keys):
    - `'boxstyle'`: `'round'`, `'square'`, `'round4'`, `'sawtooth'`, `'roundtooth'`, etc.
    - `'facecolor'`: Background color
    - `'edgecolor'`: Border color
    - `'alpha'`: Transparency
    - `'pad'`: Padding

The following plots explore several of the parameter values that are available when using texts. 

#### Font sizes

````python
# Example: Different fontsize options
fig, ax = plt.subplots(figsize=(10, 4))

sizes = ['xx-small', 'x-small', 'small', 'medium', 'large', 'x-large', 'xx-large']
for i, size in enumerate(sizes):
    ax.text(0.1, 0.9 - i*0.2, f'Font size: {size}', fontsize=size, 
            transform=ax.transAxes)

# Also show numeric sizes
for i, size in enumerate([8, 12, 16, 20], 1):
    ax.text(0.6, 0.6-i*0.1, f'Size {size}', size=size) 
ax.set_title('Fontsize Parameter Examples')
ax.axis('off')

plt.show()
````

#### Font weights

````python
# Example: Different fontweight and fontstyle options
fig, ax = plt.subplots(figsize=(10, 4))

weights = ['light', 'normal', 'bold', 'heavy']
for i, weight in enumerate(weights):
    ax.text(0.1, 0.8 - i*0.15, f'Font weight: {weight}', 
            fontweight=weight, fontsize=14)

styles = ['normal', 'italic', 'oblique']
for i, style in enumerate(styles):
    ax.text(0.6, 0.8 - i*0.15, f'Font style: {style}', 
            fontstyle=style, fontsize=14)

ax.set_title('Fontweight and Fontstyle Parameter Examples')
ax.axis('off')

plt.show()
````

#### Color and transparency

````python
# Example: Color and alpha parameters
fig, ax = plt.subplots(figsize=(10, 6))

# Named colors
colors = ['red', 'blue', 'green', 'purple', 'orange']
for i, color in enumerate(colors):
    ax.text(0.1, 0.8 - i*0.12, f'Color: {color}', color=color, 
            fontsize=14, fontweight='bold')

# Hex colors
ax.text(0.6, 0.8, 'Hex color #FF5733', color='#FF5733', 
        fontsize=14, fontweight='bold')

# Alpha transparency
alphas = [1.0, 0.7, 0.4, 0.2]
for i, alpha in enumerate(alphas):
    ax.text(0.6, 0.6 - i*0.12, f'Alpha: {alpha}', color='blue', 
            alpha=alpha, fontsize=14, fontweight='bold')

ax.set_title('Color and Alpha Parameter Examples')
ax.axis('off')

plt.show()
````

#### Alignments

````python
# Example: Horizontal and vertical alignment
fig, ax = plt.subplots(figsize=(8, 4))

# Mark the reference point
ref_x, ref_y = 0.3, 0.5

# Vertical alignments
va_options = ['top', 'center', 'bottom', 'baseline']
for i, va in enumerate(va_options):
    ax.plot(ref_x + i*0.20,  ref_y+0.15, 'ro', markersize=5, transform=ax.transAxes, alpha=0.5)
    ax.text(ref_x + i*0.20,  ref_y+0.15,
            s=f'va="{va}"',#
            ha= 'center', va=va,# 
            fontsize=10, transform=ax.transAxes,#
            bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.3))

# Horizontal alignments
ha_options = ['left', 'center', 'right']
for i, ha in enumerate(ha_options):
    ax.plot(ref_x + i*0.25, ref_y, 'ro', markersize=5, transform=ax.transAxes, alpha=0.9)
    ax.text(ref_x + i*0.25, ref_y, f'ha="{ha}"', ha=ha, va='center', transform=ax.transAxes,
            size=12, bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.1))

ax.set_title('Horizontal and Vertical Alignment Examples')
ax.axis('off')
plt.show()
````

#### Rotations

## Annotation Objects  `ax.annotate()`

The `ax.annotate()` method creates annotations with optional arrows pointing to specific locations. It's used for:
- Highlighting specific data points with explanatory text
- Drawing attention to important features in plots
- Creating clear visual connections between text and data
- Adding contextual information with directional indicators


**Main annotation Parameters:**
- `text`: The annotation text string.
- `xy`: The point (x, y) to annotate (where arrow points)
- `xytext`: Position (x, y) of the text (optional, defaults to xy)
- `xycoords`: Coordinate system for `xy`
  - Options:
    - `'data'`: Use data coordinates (default)
    - `'axes fraction'`: Fraction of axes (0-1)
    - `'figure fraction'`: Fraction of figure (0-1)
    - `'axes points'`: Points from lower left of axes
    - `'figure points'`: Points from lower left of figure
    - `'offset points'`: Offset in points from xy
- `textcoords`: Coordinate system for `xytext`. Same as `xycoords`.
- `arrowprops`: Dictionary of arrow properties (if None, no arrow)
  - Dictionary options:
    - `arrowstyle`: Style of the arrow
      - `'-'`: None (just a line)
      - `'->'`: Simple arrow
      - `'-['`: Bracket arrow
      - `'-|>'`: Fancy arrow
      - `'<->'`: Double arrow
      - `'<|-|>'`: Fancy double arrow
      - `'fancy'`: Fancy arrow (with head width/length)
      - `'simple'`: Simple arrow
      - `'wedge'`: Wedge-shaped arrow
    - `connectionstyle`: Path style
      - `'arc3'`: Simple arc (with rad parameter)
      - `'arc3,rad=0.3'`: Arc with radius
      - `'angle3'`: Three-point arc
      - `'angle'`: Angled connection
      - `'bar'`: Bar connection
    - `color`: Color of the arrow
    - linewidth` or **`lw`**: Width of arrow line
    - `alpha`: Transparency of arrow
    - `shrinkA`: Points to shrink from annotation point (xy) - for use with `arrowstyle`
    - `shrinkB`: Points to shrink from text position (xytext) - for use with `arrowstyle`
    - `shrink`: Fraction (0-1) to shrink from both ends - only for old-style arrows (no `arrowstyle`)
    - `mutation_scale`**: Size of arrow head
- `fontsize` or `size`: Size of the annotation text. Same options as `ax.text()`.
- `color` or `c`: Color of the annotation text. Same options as `ax.text()`.
- `fontweight`: Weight of the font. Same options as `ax.text()`
- `fontstyle`: Style of the font. Same options as `ax.text()`.
- `ha`, `va`: Horizontal and vertical alignment. Same options as `ax.text()`.
- `bbox`: Bounding box properties. Same options as `ax.text()`.


```{note} Shrink parameters
- Use `shrinkA` and `shrinkB` (in points) when you specify `arrowstyle`
- Use `shrink` (fraction 0-1) only with old-style arrows (no `arrowstyle`)

The following plots explore several of the parameter values that are available when using annotations. 

#### xy and xytext

````python
# Example: Basic annotation with xy and xytext
fig, ax = plt.subplots(figsize=(8, 4))

# Sample data point
x_point, y_point = 5, 5
ax.plot(x_point, y_point, 'ro', markersize=12)

# Annotation with arrow pointing to the data point
ax.annotate('This is the point\nxy=(5, 5)', 
            xy= (x_point, y_point),
            xytext= (6,7),
            size=12,
            ha= 'center',
            arrowprops=dict(arrowstyle='->', color='black', lw=2),
            bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.7))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_title('Basic Annotation: xy and xytext Parameters')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

#### Using different coordinate systems

````python
# Example: Different coordinate systems (xycoords and textcoords)
fig, ax = plt.subplots(figsize=(12, 6))

# Plot some data
x = np.linspace(0, 10, 100)
y = np.sin(x)
ax.plot(x, y, 'b-', linewidth=2)

# Example 1: Both in data coordinates
ax.annotate('data coords', xy=(2, np.sin(2)), xytext=(3, 0.5),
            xycoords='data', textcoords='data',
            arrowprops=dict(arrowstyle='->', color='red'),
            fontsize=10, bbox=dict(boxstyle='round', facecolor='pink', alpha=0.5))

# Example 2: xy in data, xytext in axes fraction
ax.annotate('axes fraction text', xy=(5, np.sin(5)), xytext=(0.7, 0.9),
            xycoords='data', textcoords='axes fraction',
            arrowprops=dict(arrowstyle='->', color='green'),
            fontsize=10, bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.5))

# Example 3: Using offset points
ax.annotate('offset points', xy=(8, np.sin(8)), xytext=(30, -30),
            xycoords='data', textcoords='offset points',
            arrowprops=dict(arrowstyle='->', color='purple'),
            fontsize=10, bbox=dict(boxstyle='round', facecolor='plum', alpha=0.5))

ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
ax.set_title('Coordinate System Examples (xycoords and textcoords)')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

#### Connection Styles

Connection styles are a bit more cryptic to descifer. Here is a list of the main parameters used:   

- `angleA`: Angle (using math convention) leaving the text point (degrees)
- `angleB`: Angle (using math convention) arriving at target point (degrees)
- `rad`: Rounds the corner (0 = sharp corner) (fraction of distance or points)
- `armA`: Length of vertical arm from text point (points)
- `armB`: Length of vertical arm to target point (points)
- `fraction`: Where the bar sits between points (0.0 = near text, 1.0 = near target)
- `angle`: Rotation angle of the entire bar structure

| Connection Style | Parameters | Description | Example |
|------------------|------------|-------------|---------|
| `arc3` | `rad` | Simple quadratic Bezier curve. `rad` controls curvature: 0=straight, positive=curves left/CCW, negative=curves right/CW | `connectionstyle='arc3,rad=0.3'` |
| `arc` | `angleA`, `angleB`, `armA`, `armB`, `rad` | Creates a path with two straight "arms" connected by a curved arc in the middle. `angleA`/`angleB`: angles of arms at text/target (degrees). `armA`/`armB`: length of arms. `rad`: arc radius | `connectionstyle='arc,angleA=0,angleB=90,armA=30,armB=30,rad=5'` |
| `angle3` | `angleA`, `angleB` | Simple angled connector using two straight segments. `angleA`: angle at text point. `angleB`: angle at target point | `connectionstyle='angle3,angleA=90,angleB=0'` |
| `angle` | `angleA`, `angleB`, `rad` |Creates a strict right-angle (90°) bend between two straight segments, with an optional rounded corner.. `angleA`/`angleB`: angles at text/target. `rad`: corner rounding radius | `connectionstyle='angle,angleA=0,angleB=90,rad=10'` |
| `bar` | `armA`, `armB`, `fraction`, `angle` |Creates a bracket or "Z" shaped connector with a bar segment between two arms.. `armA`/`armB`: arm lengths. `fraction`: position of bar between points (0-1). `angle`: bar angle in degrees | `connectionstyle='bar,armA=10,armB=10,fraction=0.3,angle=180'` |


````python
# Example: Different connectionstyles
fig, ax = plt.subplots(figsize=(12, 8))

connectionstyles = [
    'arc3',
    'arc3,rad=0.1',
    'arc, angleA=135,angleB=45,armA=60,armB=50,rad=5',
    'angle3,angleA=-30,angleB=30',
    'angle, angleA=-10, angleB=10, rad=3',
    'bar, armA=0, armB=0, fraction=-0.1'
]

for i, style in enumerate(connectionstyles):
    y_pos = 7 - i * 1.2
    ax.plot(2, y_pos, 'ro', markersize=8)
    
    ax.annotate(f'connectionstyle="{style}"',
                xy=(2, y_pos),
                xytext=(6, y_pos),
                fontsize=9,
                arrowprops=dict(arrowstyle='->', 
                                connectionstyle=style,
                                color='darkgreen', lw=2),
                bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.6))

ax.set_xlim(0, 10)
ax.set_ylim(0, 8)
ax.set_title('Connection Style (connectionstyle) Parameter Examples')
ax.axis('off')

plt.show()
````

#### Arrow Properties

````python
# Example: Arrow properties (color, linewidth, alpha, shrinkA/shrinkB)
fig, ax = plt.subplots(figsize=(12, 8))

# Different colors
colors = ['red', 'blue', 'green', 'purple', 'orange']
for i, color in enumerate(colors):
    y_pos = 7 - i * 0.7
    ax.plot(2, y_pos, 'ko', markersize=6)
    ax.annotate(f'color="{color}"', xy=(2, y_pos), xytext=(4, y_pos),
                arrowprops=dict(arrowstyle='->', color=color, lw=2),
                fontsize=10)

# Different linewidths
linewidths = [1, 2, 3, 4]
for i, lw in enumerate(linewidths):
    y_pos = 3.5 - i * 0.8
    ax.plot(2, y_pos, 'ko', markersize=6)
    ax.annotate(f'linewidth={lw}', xy=(2, y_pos), xytext=(4, y_pos),
                arrowprops=dict(arrowstyle='->', color='steelblue', lw=lw),
                fontsize=10)

# Different alpha values
alphas = [1.0, 0.7, 0.4, 0.2]
for i, alpha in enumerate(alphas):
    x_pos = 7 + i * 1.5
    ax.plot(x_pos, 7, 'ko', markersize=6)
    ax.annotate(f'alpha={alpha}', xy=(x_pos, 7), xytext=(x_pos, 5.5),
                arrowprops=dict(arrowstyle='->', color='red', lw=3, alpha=alpha),
                fontsize=9, ha='center')

# shrinkA and shrinkB parameters
shrinks = [0, 5, 10, 15]
for i, shrink in enumerate(shrinks):
    x_pos = 7 + i * 1.5
    ax.plot(x_pos, 3.0, 'ko', markersize=10)
    ax.annotate(f'shrink={shrink}', xy=(x_pos, 3.0), xytext=(x_pos, 1.5),
                arrowprops=dict(arrowstyle='->', color='green', lw=2,
                shrinkA= shrink, shrinkB=shrink),
                fontsize=9, ha='center')

ax.set_xlim(0, 13)
ax.set_ylim(0, 8)
ax.set_title('Arrow Properties: color, linewidth, alpha, shrinkA/shrinkB')
ax.axis('off')
plt.show()
````

````python
# Annotating outliers in scatter plot
fig, ax = plt.subplots(figsize=(8, 8))

# create random data
np.random.seed(2026)
n_points = 100
x = np.random.randn(n_points) * 2 + 5
y = x * 1.5 + np.random.randn(n_points) * 2 + 3

# create outliers
outlier_x = [2, 12, 8]
outlier_y = [15, 3, 20]

# add outliers to data
x = np.append(x, outlier_x)
y = np.append(y, outlier_y)

# Plot Data
ax.scatter(x[:-3], y[:-3], alpha=0.5, s=50, c='steelblue', label='Normal data') # regular points
ax.scatter(outlier_x, outlier_y, s=150, c='red', marker='X',
           edgecolors='darkred', linewidths=1, label='Outliers') # outliers

# Annotate each outlier
outlier_reasons = [
    'Measurement\nerror',
    'Equipment\nmalfunction',
    'Unusual\nevent'
]

for i, (ox, oy, reason) in enumerate(zip(outlier_x, outlier_y, outlier_reasons)):
    ax.annotate(reason,
                xy=(ox, oy),
                xytext=(-60, 30),
                textcoords='offset points',
                fontsize=10, fontweight='bold',
                arrowprops=dict(arrowstyle='-|>', color='red', lw=2,
                               connectionstyle='arc3,rad=0.2'),
                bbox=dict(boxstyle='round,pad=0.5', facecolor='yellow', 
                         edgecolor='red', linewidth=2, alpha=0.9))

ax.set_xlabel('X Variable', fontsize=12)
ax.set_ylabel('Y Variable', fontsize=12)
ax.set_title('Scatter Plot with Outlier Annotations', fontsize=14, fontweight='bold')
ax.legend(loc='upper left')
ax.grid(True, alpha=0.3)
plt.show()
````

#### `ax.text()` .vs. `ax.annotate()`

**Choose `ax.text()` when:**

- You need simple labels without arrows
- Labeling axes, adding titles, or placing notes
- Space allows direct labeling near data
- Working with dense data where arrows would clutter
- Performance is critical (many labels)

**Choose `ax.annotate()` when:**

- You need to point to specific data points
- Data is crowded and text must be offset
- Highlighting outliers or special values
- Need visual guidance from text to data
- Explaining specific features in the plot

#### General Guidelines:

- **Don't over-annotate**: Too many annotations clutter the plot
- **Use consistent styling**: Keep fonts, colors, and box styles uniform
- **Consider readability**: Ensure text size is appropriate for the figure size
- **Avoid overlaps**: Position text and arrows to not obscure data
- **Match arrow colors to context**: Use meaningful colors (red for warnings, green for success)
