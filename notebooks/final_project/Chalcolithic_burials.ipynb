```python
import pandas as pd
dfage = pd.read_excel(file_path, sheet_name=0)  # or "SheetName"
dfage = dfage.rename(columns={"Sub-Region": "sub_region"})
```

```python
import matplotlib.pyplot as plt
import numpy as np
#%matplotlib inline 
# Set random seed for reproducibility
np.random.seed(2026)

import plotnine as pln
from plotnine.data import penguins 

#p = ggplot(anscombe_quartet, aes(x="x", y="y")) + geom_point()
from plotnine.data import diamonds
#aes("color", fill="cut")
p = pln.ggplot(dfage, pln.aes("sub_region", "Age", fill=('Burials'))) + pln.theme(legend_position="none")

p + pln.geom_bar(position=pln.position_stack())  # default
p + pln.geom_bar(position=pln.position_fill()) #equalizes heigh
p + pln.geom_bar(position=pln.position_dodge())
# come back and facet these together
```
